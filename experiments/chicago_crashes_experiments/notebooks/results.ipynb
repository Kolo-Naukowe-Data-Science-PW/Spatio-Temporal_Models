# Chicago crashes: model comparison
Load aggregated metrics and compare MAE/RMSE across models.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

metrics = pd.read_csv("../outputs/metrics.csv")
if "split" in metrics.columns:
    metrics = metrics[metrics["split"].eq("spatiotemporal_cv_mean")].copy()
metrics
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 4))
ax = sns.barplot(data=metrics, x="model", y="rmse")
ax.bar_label(ax.containers[0], fmt="%.2f")
plt.title("RMSE by model")
plt.xlabel("")
plt.ylabel("RMSE")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
ax = sns.barplot(data=metrics, x="model", y="mae")
ax.bar_label(ax.containers[0], fmt="%.2f")
plt.title("MAE by model")
plt.xlabel("")
plt.ylabel("MAE")
plt.tight_layout()
plt.show()
```
